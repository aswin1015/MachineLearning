```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")
```

```python
fre=10

sample_fre=100

t=np.linspace(0,2,2*sample_fre,)
a=np.sin(fre*2*np.pi*t)
plt.plot(t,a)
plt.ylabel("Simple amplitude")
plt.xlabel("Time(s)")
plt.show()
```

```python
from scipy import fftpack
```

```python
A= fftpack.fft(a)
frequency = fftpack.fftfreq(len(a))*sample_fre

plt.stem(frequency,np.abs(A),use_line_collection=True)
plt.xlabel("Frequency in Hz")
plt.ylabel("Frequency Spectrum in magnitude")
plt.show()
```
